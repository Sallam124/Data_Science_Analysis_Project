## Statistical Analysis
- Conduct tests such as t-tests, ANOVA, and chi-squared to explore relationships.

```python
action_sales = data3[data3['Genre'] == 'Action']['Global_Sales_Millions']
shooter_sales = data3[data3['Genre'] == 'Shooter']['Global_Sales_Millions']

T_test_shooter_action = pd.DataFrame({
    'Genre': ['Action'] * len(action_sales) + ['Shooter'] * len(shooter_sales),
    'Global_Sales_Millions': list(action_sales) + list(shooter_sales)
})

# Create the box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x='Genre', y='Global_Sales_Millions', data=T_test_shooter_action)
plt.title('T-Test: Action vs. Shooter Genres')
plt.xlabel('Genre')
plt.ylabel('Global Sales (in millions)')
plt.show()
```

```python
# Group by Genre and get Global Sales data
ANOVA_Genre_Global_Sales = data3.groupby('Genre')['Global_Sales_Millions'].apply(list)

ANOVA_Genre_Global_Sales_result = stats.f_oneway(*ANOVA_Genre_Global_Sales)

print("ANOVA test result for Global Sales by Genre:")
print(f"F-statistic: {ANOVA_Genre_Global_Sales_result.statistic}")
print(f"P-value: {ANOVA_Genre_Global_Sales_result.pvalue}")

plt.figure(figsize=(10, 6))
sns.boxplot(x='Genre', y='Global_Sales_Millions', data=data3)
plt.title('Global Sales Comparison by Genre (ANOVA Test)')
plt.xlabel('Genre')
plt.ylabel('Global Sales (in millions)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a table for Platform and Genre
Chi2_Platform_Genre_Table = pd.crosstab(data3['Platform'], data3['Genre'])

# Perform the Chi-Squared test
Chi2_Platform_Genre_result = stats.chi2_contingency(Chi2_Platform_Genre_Table)

print("Chi-Squared Test result for Platform vs Genre:")
print(f"Chi-squared statistic: {Chi2_Platform_Genre_result[0]}")
print(f"P-value: {Chi2_Platform_Genre_result[1]}")

# Visualize the table as a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(Chi2_Platform_Genre_Table, annot=True, cmap='YlGnBu', fmt='d')
plt.title('Platform vs Genre Contingency Table (Chi-Squared Test)')
plt.xlabel('Genre')
plt.ylabel('Platform')
plt.show()
```

```python
# Group by Release Season and get Global Sales data
ANOVA_Release_Season_Global_Sales = data3.groupby('Release_Season')['Global_Sales_Millions'].apply(list)

ANOVA_Release_Season_Global_Sales_result = stats.f_oneway(*ANOVA_Release_Season_Global_Sales)

print("ANOVA test result for Global Sales by Release Season:")
print(f"F-statistic: {ANOVA_Release_Season_Global_Sales_result.statistic}")
print(f"P-value: {ANOVA_Release_Season_Global_Sales_result.pvalue}")

# Create the box plot for visualization
plt.figure(figsize=(10, 6))
sns.boxplot(x='Release_Season', y='Global_Sales_Millions', data=data3)
plt.title('Global Sales Comparison by Release Season (ANOVA Test)')
plt.xlabel('Release Season')
plt.ylabel('Global Sales (in millions)')
plt.show()
```

```python
ANOVA_Year_of_Release_Global_Sales = data3.groupby('Year_of_Release')['Global_Sales_Millions'].apply(list)

ANOVA_Year_of_Release_Global_Sales_result = stats.f_oneway(*ANOVA_Year_of_Release_Global_Sales)

print("ANOVA test result for Global Sales by Year of Release:")
print(f"F-statistic: {ANOVA_Year_of_Release_Global_Sales_result.statistic}")
print(f"P-value: {ANOVA_Year_of_Release_Global_Sales_result.pvalue}")

plt.figure(figsize=(12, 6))
sns.boxplot(x='Year_of_Release', y='Global_Sales_Millions', data=data3)
plt.title('Global Sales Comparison by Year of Release (ANOVA Test)')
plt.xlabel('Year of Release')
plt.ylabel('Global Sales (in millions)')
plt.xticks(rotation=45)
plt.show()
```

```python

from scipy.stats import pearsonr

# Convert User_Score to numeric, forcing errors to NaN (if any)
data3['User_Score'] = pd.to_numeric(data3['User_Score'], errors='coerce')

# Drop rows where User_Score or Global_Sales_Millions is NaN
filtered_data = data3.dropna(subset=['User_Score', 'Global_Sales_Millions'])

# Calculate the Pearson correlation
correlation, p_value = pearsonr(filtered_data['User_Score'], filtered_data['Global_Sales_Millions'])

print(f"Pearson Correlation between User Scores and Global Sales: {correlation}")
print(f"P-value: {p_value}")

# Plot the scatter plot with a regression line
plt.figure(figsize=(8, 6))
sns.regplot(x='User_Score', y='Global_Sales_Millions', data=filtered_data, scatter_kws={'s': 10}, line_kws={'color': 'red'})
plt.title('User Scores vs Global Sales')
plt.xlabel('User Score')
plt.ylabel('Global Sales (in millions)')
plt.show()
```
